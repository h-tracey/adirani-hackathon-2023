```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("per page with type.csv")
# df.drop("index", axis="columns", inplace=True)
df['ptype'] = pd.Categorical(df['ptype'])
```

```python
df.dropna(subset=['ptype']).plot.hexbin('conversion_rate_per_user', 'views_per_user', gridsize=40)
```

```python
df.dropna(subset=['ptype']).plot.scatter('conversion_rate_per_user', 'views_per_user', c='ptype', cmap='viridis')
```

```python
df.boxplot(['views_per_user'], by='ptype')
```

```python
df.boxplot('conversion_rate_total', by='ptype')
```
